```python
# Import the Open Data Cube and create an instance of the class
import datacube
dc = datacube.Datacube()
```

```python
# We want to find all the available linescan datasets in the `linescan` product
product = "linescan"
linescan_datasets = dc.find_datasets(product=product)
print(f"Found {len(linescan_datasets)} datasets that can be loaded")
```

```python
# Load the dataset at the listed index
dataset_index = 0
ds = dc.load(product=product, id=linescan_datasets[dataset_index].id, output_crs="epsg:28355", resolution=(-20, 20))
```

```python
# Do a simple plot of the data to confirm we can load it
ds.thermal.plot(size=8)
```
